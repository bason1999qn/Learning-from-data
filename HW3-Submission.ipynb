## Câu 1 (1 điểm)

Generalization Error ám chỉ $|E_{in}(g) - E_{out}(g)|$.

Dựa vào đề bài mô tả, ta cần giải bất phương trình sau (theo $N$):

$2Me^{-2\times0.05^2N} \leq 0.03$

$\Leftrightarrow N \geq$ $\frac{ln(\frac{0.03}{2M})}{-2 * 0.05^{2}}$
**(Công thức dùng chung cho câu 1,2 và 3)**


<br>Tại M = 1, ta có:
<br>$\Leftrightarrow N \geq$ $\frac{ln(\frac{0.03}{2*1})}{-2 * 0.05^{2}}$

```python
import numpy as np

def computeN(M):
    return - np.log(0.03/(2*M))/(2*0.05**2)
M = 1
print("N >= ", computeN(M))
```

- iv) $2^{\lfloor \sqrt{N/2} \rfloor}$: đây không có break point và cũng không phải đa thức, do đó nó không phải là hàm tăng trưởng (growth function).
